```python
transactions=[]
for productId in mergedData[mergedData.cluster==2].groupby("order_id")["product_id"]:
    transactions.append(list(productId[1]))
```

```python
transactions
```

```python
from apyori import apriori
associationRules=list(apriori(transactions, min_support=0.003, min_confidence=0.2, min_lift=2, max_length =3))
# associationResults=list(association_rules)
```

```python
for item in associationRules:

    # first index of the inner list
    # Contains base item and add item
    pair = item[0] 
    items = [x for x in pair]
    print("Rule: " + str(products[products.product_id==items[0]].product_name.unique()[0]) + " -> " + str(products[products.product_id==items[1]].product_name.unique()[0]))

    #second index of the inner list
    print("Support: " + str(item[1]))

    #third index of the list located at 0th
    #of the third index of the inner list

    print("Confidence: " + str(item[2][0][2]))
    print("Lift: " + str(item[2][0][3]))
    print("=====================================")
```
